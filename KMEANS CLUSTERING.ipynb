## Dataset 2 (bike-sharing.csv)

```python
import pandas as pd #import library

df_2 = pd.read_csv(("bike-sharing.csv"), index_col = "instant") #input data csv
df_2.head(3) #menampilkan 3 data teratas
```

```python
import matplotlib.pyplot as plt #import library
import seaborn as sns #import library

#membuat plot
plt.figure(1, figsize = (10, 5)) #ukuran plot
sns.scatterplot( data= df_2, #data yang ingin divisualisasi
                x = "cnt", #label x
                y = "registered", #label y
                hue = "holiday", # warna berdasarkan holiday
                size = "yr", #ukuran berdasarkan tahun/year
                palette = "Set2" #palette warna plot
);
```

```python
from sklearn import preprocessing #import library

X2 = df_2.drop(["dteday",	"season", "yr", "mnth", "hr",	"holiday",	"weekday",
                  "workingday",	"weathersit","temp",	"atemp"	,"hum", "windspeed",
                  "casual"], axis = 1) #menghapus kolom yang tidak diperlukan
X2_norm = preprocessing.normalize(X2) #normalisasi data
```

```python
import numpy as np #import library
from sklearn.cluster import KMeans #import library

#fungsi elbow_plot
def elbow_plot(data,clusters): #definisikan fungsi
  inertia = [] #definisikan variabel inertia
  for n in range(1, clusters): #membuat perulangan dengan parameter(1 dan clusters)
      algorithm = KMeans(
        n_clusters=n, init="k-means++", random_state=125,
     ) #membuat algoritma k-means, dengan memasukan jumlah kluster, initialisasi dan random state
      algorithm.fit(data) #dilakukan fitting algoritma dari data
      inertia.append(algorithm.inertia_) #penambahan inertia pada algoritma

#menghasilkan plot
  plt.plot(np.arange(1, clusters), inertia, 'o') #membuat plot dan label
  plt.plot(np.arange(1, clusters), inertia, '-', alpha = 0.5 )#membuat plot dan label
  plt.xlabel("Number of Clusters"), plt.ylabel('Inertia') #membuat label x dan y
  plt.show(); #menunjukkan output
```

```python
#membuat algoritma untuk clustering k-means
algorithm = KMeans(n_clusters = 3, init = 'k-means++', random_state = 125)
#algoritma k-means dengan jumlah cluster 3 dan random state 125
algorithm.fit(X2_norm) #memasukkan data yang telah di normalisasi kedalam fitting algoritma
labels = algorithm.labels_ #membuat label dari algoritma
```

```python
#plot hasil clustering k-means
sns.scatterplot(data = X2, x= 'cnt', y = 'registered', hue = labels, palette = "Set2");
```

```python
#mencari titik k optimal dari fungsi elbow yang telah dibuat diatas
elbow_plot(X2_norm,10)
```

```python
  # visualisasi menggunakan boxplot
  sns.boxplot(x = labels, y= X2['cnt'])
```
